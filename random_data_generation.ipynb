```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.0, 10.0, num=100)
a = 4.0
b = -3.5
c = 0.0

y = (a*(x**2)) + (b*x) + c 

#   let's add noise to the data
#   np.random.normal(mean, standardDeviation, num)
noise = np.random.normal(0, 10., 100)
y = y+noise
x = np.arange(0, 100, step = 1)

plt.scatter(x, y)
plt.show()
```
